```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
crimes = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery']
offenses_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
change_in_offenses = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]
```

```python
sorted_crimes, sorted_offenses = zip(*sorted(zip(crimes, offenses_2023), key=lambda x: x[1], reverse=True))

plt.figure(figsize=(10, 6))
bars = plt.bar(sorted_crimes, sorted_offenses)

for index, value in enumerate(sorted_offenses):
    plt.text(index, value + 10, f'{value:.1f}', ha='center', color='black', fontweight='bold')

plt.xlabel('Type of Crime', fontsize=12, fontweight='bold')
plt.ylabel('Offenses per 100,000 People', fontsize=12, fontweight='bold')
plt.title('Crime Rates in 2023 (Offenses per 100,000 People)', fontsize=14, fontweight='bold')
plt.grid(True, axis='y', linestyle='--', linewidth=0.7)

plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python
Offense=['Aggravated Assault','Burglary','Larceny','Motor Vehicle Theft','Murder','Rape (revised definition)','Robbery']
# Rate_2023=[264.1,250.7,1347.2,318.7,5.7,38.0,66.5]
Rate_Change_22_23=[-3.3,-8.1,-4.9,12.0,-12.0,-9.8,-0.8]

plt.figure(figsize=(12, 7))

#assign unique color to each crime type
rainbow = ['#D2B3CB', '#432E70', '#007991', '#439A86', '#E9D985', '#F7A072', '#F45B69']

plt.barh(Offense, Rate_Change_22_23, color=rainbow, zorder=2)

#label axes, add title
plt.ylabel('Offense')
plt.xlabel('Percent Change (2022-2023)')
plt.title('Change in Offense Rates (2022-2023)')

plt.grid(axis='x', zorder=0);
```

```python
data = {'Offense' : ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery'],
        'Percent Change' : [-3.3,-8.1,-4.9,12.0,-12.0,-9.8,-0.8]}

dm = pd.DataFrame(data)
rainbow = ['#D1B1CB', '#444E70', '#007991', '#439A86', '#E9D824', '#F7A072', '#F45C48']

dm_sorted = dm.sort_values(by='Percent Change', ascending=True)

plt.figure(figsize=(12, 7))

plt.barh(dm_sorted['Offense'], dm_sorted['Percent Change'], color=rainbow, zorder=2)

plt.ylabel('Offense')
plt.xlabel('Percent Change in Offenses per 100,000 People (2022-2023)')
plt.title('Change in Crime Rates (2022-2023)')

ax = plt.gca()
ax.tick_params(axis='x', labelsize=8)
ax.tick_params(axis='y', labelsize=8)
plt.grid(axis='x', linestyle='-', color='#dddddd', zorder=0)

plt.show()
```
